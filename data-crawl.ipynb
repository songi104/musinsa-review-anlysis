```python
from google_play_scraper import reviews
from datetime import datetime
import pandas as pd
```

```python

result, continuation_token = reviews("com.musinsa.store", lang="ko", country="kr")
cnt = 0

while True:
    cnt += 1
    print(f"{cnt}번째 크롤링 중...")
    tlst, continuation_token = reviews("com.musinsa.store", lang="ko", country="kr", continuation_token=continuation_token)
    result += tlst
    last = result[-1]["at"]
    
    if cnt == 19:
        break
```
