```python
# python 3.12
import pandas as pd
from konlpy.tag import Okt
from collections import Counter
```

```python
path = "250205_review_data.xlsx"
df = pd.read_excel(path)
```

```python
df.head()
```

```python
# 결측지 확인
print(df.isnull().sum())
```

```python
# 불필요한 열 제거
df = df.drop('Unnamed: 0', axis=1)
```

```python
import os
```

```python
df = df.drop(['content_nouns', 'content_words'], axis=1)
```

```python
# from konlpy.tag import Okt
okt = Okt()
df['content_words'] = df['content'].apply(lambda x:list(set(okt.nouns(x))))
df.head()
```

```python
df.head()
```

```python
counterByScore = {}
for i in range(1, 6):
    t_df = df[df['score'] == i]
    all_words = []
    for words, count in zip(t_df['content_words'], t_df['thumbsUpCount']):
        all_words.extend(words * int(count))

    words_counter = Counter(all_words)
    counterByScore[i] = words_counter
```

```python
counterByScore[1]
```

```python
count_data = []

for score, counter in counterByScore.items():
    for word, count in counter.items():
        count_data.append({'score': score, 'word': word, 'count': count})

count_df = pd.DataFrame(count_data)
```

```python
date = path[:6]
count_df.to_excel(date+'_review_word_count.xlsx')
```
