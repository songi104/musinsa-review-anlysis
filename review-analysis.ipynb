```python
# python 3.12
import pandas as pd
from konlpy.tag import Okt
from collections import Counter
```

```python
path = "250208_review_data.xlsx"
df = pd.read_excel(path)
```

```python
df.head()
```

```python
# 결측지 확인
print(df.isnull().sum())
```

```python
# 불필요한 열 제거
df = df.drop('Unnamed: 0', axis=1)
```

```python
def preprocess(string):
    okt = Okt()
    lst = okt.pos(string, stem=True, norm=True)
    res = []
    for word, pumsa in lst:
        if pumsa in ["Noun", "Adjective", "Verb"]:
            res.append(word)
    return res
```

```python
# 10개정도 꺼내서 test
okt = Okt()
temp_df = df.iloc[:5]
# temp_df = temp_df.drop(['at', 'reviewId'], axis=1)
temp_content = df.iloc[0]['content']
# print(okt.pos(temp_content, stem=True, norm=True))
temp_df['content_words'] = temp_df['content'].apply(lambda x:preprocess(x))
temp_df.head()
```

```python
# from konlpy.tag import Okt
okt = Okt()
df['content_words'] = df['content'].apply(lambda x:preprocess(x))
df.head()
```

```python
df.head()
```

```python
counterByScore = {}
for i in range(1, 6):
    t_df = df[df['score'] == i]
    all_words = []
    for words, count in zip(t_df['content_words'], t_df['thumbsUpCount']):
        all_words.extend(words * int(count))

    words_counter = Counter(all_words)
    counterByScore[i] = words_counter
```

```python
counterByScore[1]
```

```python
count_data = []

for score, counter in counterByScore.items():
    for word, count in counter.items():
        count_data.append({'score': score, 'word': word, 'count': count})

count_df = pd.DataFrame(count_data)
```

```python
date = path[:6]
count_df.to_excel(date+'_review_word_count.xlsx')
```
