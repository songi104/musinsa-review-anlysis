```python
# python 3.12
import pandas as pd
from konlpy.tag import Okt
from collections import Counter
```

```python
path = "250205_review_data.xlsx"
df = pd.read_excel(path)
```

```python
df.head()
```

```python
# 결측지 확인
print(df.isnull().sum())
```

```python
# 불필요한 열 제거
df = df.drop('Unnamed: 0', axis=1)
```

```python
import os
```

```python
# from konlpy.tag import Okt
okt = Okt()
df['content_nouns'] = df['content'].apply(lambda x:okt.morphs(x))
df.head()
```

```python
# '말해보카' 라는 것을 말, 카로 split하는 문제 해결 필요
```

```python
df.head()
```

```python
counterByScore = {}
for i in range(1, 6):
    t_df = df[df['score'] == i]
    all_nouns = sum(t_df['content_nouns'].tolist(), [])
    nouns_counter = Counter(all_nouns)
    counterByScore[i] = nouns_counter
```

```python
counterByScore[1]
```

```python
count_data = []

for score, counter in counterByScore.items():
    for word, count in counter.items():
        count_data.append({'score': score, 'word': word, 'count': count})

count_df = pd.DataFrame(count_data)
```

```python
date = path[:6]
count_df.to_excel(date+'_review_word_count.xlsx')
```
